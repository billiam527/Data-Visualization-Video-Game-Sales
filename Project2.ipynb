### Import Necessary Libraries
We will need to import csv, numpy, pandas, matplotlib, seaborn and maplotlib.plyplot.  We will refer to them as np, pd, sns, mtplt and plt from here on.

```python
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mtplt
import seaborn as sns
```

### Import and Explore the Data
Import the data using read_csv. Then, read the data into "data."  Once we have our data, use the built-in methods to look at the data with various views.

```python
data = pd.read_csv('Video_Games_Sales_as_at_22_Dec_2016.csv')
data
```

```python
df1 = data.set_index('Platform')
df1
```

### Data Manipulation and Visualization
Next, we manipulate the data so that it will align with what we want to show with our visualizations. Then, we will formulate our graphics.

```python
df2 = df1.groupby('Platform').sum()
df2
```

```python
df3 = df2.iloc[:,1:6]
df3
```

```python
df4 = df3.sort_values('Global_Sales',ascending=False)
df5 = df4.head(15)
df5
```

### Data Visualizations 
First, we will explore the global sales of each of the systems.  We will do this witha stacked bar chart, a pie chart, and an area chart.

```python
df5.plot(kind='bar', stacked=True, colormap = "cool")

plt.show()
```

```python
df5.plot(kind = "pie", y = "Global_Sales", legend = False, colormap ="CMRmap")

plt.show()
```

```python
df5.plot.area(stacked=False)

plt.show()
```

We can also take a look at specific consoles.  In the case below, we will look at Wii, Xbox 360, and PS3 sales.

```python
PS3_Sales = df1.loc["PS3"]
PS3_Sales = PS3_Sales[['Year_of_Release','Global_Sales']]
PS3_Sales = PS3_Sales.groupby('Year_of_Release').sum()

X360_Sales = df1.loc["X360"]
X360_Sales = X360_Sales[['Year_of_Release','Global_Sales']]
X360_Sales = X360_Sales.groupby('Year_of_Release').sum()

Wii_Sales = df1.loc["Wii"]
Wii_Sales = Wii_Sales[['Year_of_Release','Global_Sales']]
Wii_Sales = Wii_Sales.groupby('Year_of_Release').sum()

PS3_Sales.plot(kind='bar', color = "b", title = "PS3 Sales by Year")
X360_Sales.plot(kind='bar', color = "g", title = "Xbox 360 Sales by Year")
Wii_Sales.plot(kind='bar', color = "r", title = "Wii Sales by Year")

plt.tight_layout()
plt.show()
```

Finally, let's take a look at the correlation between user scores and critic scores as well as their correlations with global sales.

```python
df1['User_Score'] = df1['User_Score'].convert_objects(convert_numeric= True)

sns.jointplot(x='Critic_Score',y='User_Score',data=df1 ,kind='hex')

plt.show()
```

```python
sns.regplot(x="Critic_Score", y="Global_Sales", data=df1, color = 'green')

plt.show()
```

```python
sns.regplot(x="User_Score", y="Global_Sales", data=df1, color = 'green')

plt.show()
```
