```python
df1 = data.set_index('Platform')
df1
```

### Data Manipulation and Visualization
Next, we manipulate the data so that it will align with what we want to show with our visualizations. Then, we will formulate our graphics.

```python
df2 = df1.groupby('Platform').sum()
df2
```

```python
df3 = df2.iloc[:,1:6]
df3
```

```python
df4 = df3.sort_values('Global_Sales',ascending=False)
df5 = df4.head(15)
df5
```

```python
df5.plot(kind='bar', stacked=True)

plt.show()
```

```python
df6 = df1.iloc[:,8:].groupby('Platform').mean()

df6
```

```python
df6.plot(kind="scatter", x="Global_Sales", y="Critic_Score")

plt.show()
```
